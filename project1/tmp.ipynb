```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"results\experiment01_results.csv")
# r2 = pd.read_csv(r"results\experiment01_results2.csv").groupby(["Name", "Epoch"])[["loss", "accuracy", "val_loss", "val_accuracy"]].mean().reset_index()
# r3 = pd.read_csv(r"results\experiment01_results3.csv").groupby(["Name", "Epoch"])[["loss", "accuracy", "val_loss", "val_accuracy"]].mean().reset_index()
```
